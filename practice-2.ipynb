```python
#load the needed packages and modules 
import pandas as pd
import numpy as np
from datetime import date
import datetime as dt
```

```python
import boto3
s3_client = boto3.client('s3')

```

```python
import io
obj = s3_client.get_object(Bucket='my-practice-bucket-hvz', Key='project_data.csv')
# This should prevent the 2GB download limit from a python ssl internal
chunks = (chunk for chunk in obj["Body"].iter_chunks(chunk_size=1024**3))
data = io.BytesIO(b"".join(chunks)) # This keeps everything fully in memory
df = pd.read_csv(data) # here you can provide also some necessary args and kwargs
```

```python
#since the widow and widowed are the same data so we should replace one of them using replace method
df['marital_status']=df['marital_status'].replace('Widow','Widowed')
```

```python
# we can also get the unique value for the categorical variables as well as there count to get the proportion of each category
df['marital_status'].value_counts()
```

```python
#to check the number of duplicate values in the customer id column since it is a primary field
df['customer_id'].duplicated().sum()
```

```python
#select and filter data using loc and iloc
#lets ssay we want to take a subset of the data with some columns so we can do by passing a list of columns
subset_data=df[['customer_id','educational_level','annual_income']]
subset_data.head()
```

```python
#lets say if we wanna get those rows where the educational level of the candidates is PhD
df[df['educational_level']=='PhD']
```

```python
#there is another method known as loc and iloc which we can use to filter the data
#so lets say i want to get all the rows but the columns i want is educational level marital status and annual income
#loc is an label based filtering method and the iloc is a integer position based filtering method
df.loc[:,['educational_level','marital_status','annual_income']]
```

```python
#using iloc 
df.iloc[:,[2,3,4]]
```

```python
#using loc and isin
#lets say we want to get only those rows where the educational level is graduation and master
df.loc[df['educational_level'].isin(['Graduation','Master']),:]
```

```python
#using loc and isin
#lets say we want to get only those rows where the educational level is graduation and master and i want only the customer id ,year of birth and educaional level and annual income in the columns
df.loc[df['educational_level'].isin(['Graduation','Master']),['customer_id','educational_level','marital_status','annual_income']]
```
